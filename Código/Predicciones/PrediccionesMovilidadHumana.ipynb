### Feature Engineering

**Disntancia**

Como las coordenadas iniciales y finales son conocidad, la distancia entre ambos puntos puede ser calculada utilizando la fórmula de Haversine. Esta formula nos determina la 'great-circle' distance, la cual es la distancia más corta entre dos puntos en la superficie de una esfera, medida a lo largo de la superficie de la esfera (a diferencia de una línea recta a través del interior de la esfera).

$a = \sin^{2}(\frac{\phi_{2} - \phi_{1}}{2}) + \cos(\phi_{1})\cos(\phi_{2})\sin^{2}(\frac{\lambda_{2} - \lambda_{1}}{2})$

$ d = 2ra\tan(\sqrt(\frac{a}{1-a}))$

donde $\phi$ es la latitud, $\lambda$ es la longitud, d es la distancia entre dos puntos, y r es el radio de la esfera, en nuestro caso, será reemplazado por el radio de la Tierra (6371 km) 

```python
# 1. Creamos una función que introduzca la formula de Haversine y calculamos la distancia utilizando las coordinadas de inici y fin

def haversine(start_long, start_lat, end_long, end_lat):
    
    '''
    Calculo de la distancia entre dos puntos en km 
    entre dos coordenadas dadas sus latitudes y longitudes
    '''
    
    EARTH_RADIUS = 6371 #6,371 km is the approximate distance from Earth's center to its surface 
    start_long = np.radians(start_long)
    start_lat = np.radians(start_lat)
    end_long = np.radians(end_long)
    end_lat = np.radians(end_lat)

    dlong = end_long - start_long
    dlat = end_lat - start_lat

    a = (np.sin(dlat / 2)**2 + np.cos(start_lat) * np.cos(end_lat) *
         np.sin(dlong / 2)**2)
    c = 2 * np.arctan2(np.sqrt(a), np.sqrt(1 - a))

    return np.nan_to_num(c * EARTH_RADIUS)

data["DISTANCE"] = haversine(data['START_LONG'],data['START_LAT'], data['END_LONG'], data['END_LAT'])
```

```python
input0 = Input(shape=(1,),name='p5_13_input')
print(K.shape(input0))
p5_13 = Embedding(metadata['p5_13'], 10)(input0)
p5_13 = Reshape((10,))(p5_13)
print(K.shape(p5_13))

input1 = Input(shape=(1,),name='estrato_input')
print(K.shape(input1))
estrato = Embedding(metadata['estrato'], 10)(input1)
estrato = Reshape((10,))(estrato)
print(K.shape(estrato))

input2 = Input(shape=(1,),name='sexo_input')
print(K.shape(input2))
sexo = Embedding(metadata['sexo'], 10)(input2)
sexo = Reshape((10,))(sexo)
print(K.shape(sexo))

input3 = Input(shape=(1,),name='edad_input')
print(K.shape(input3))
edad = Embedding(metadata['edad'], 10)(input3)
edad = Reshape((10,))(edad)
print(K.shape(edad))

input4 = Input(shape=(1,),name='p5_14_01_input')
print(K.shape(input4))
p5_14_01 = Embedding(metadata['p5_14_01'], 10)(input4)
p5_14_01 = Reshape((10,))(p5_14_01)
print(K.shape(p5_14_01))

input5 = Input(shape=(1,),name='p5_14_02_input')
print(K.shape(input5))
p5_14_02 = Embedding(metadata['p5_14_02'], 10)(input5)
p5_14_02 = Reshape((10,))(p5_14_02)
print(K.shape(p5_14_02))

input6 = Input(shape=(1,),name='p5_14_05_input')
print(K.shape(input6))
p5_14_05 = Embedding(metadata['p5_14_05'], 10)(input6)
p5_14_05 = Reshape((10,))(p5_14_05)
print(K.shape(p5_14_05))

input7 = Input(shape=(1,),name='p5_14_14_input')
print(K.shape(input7))
p5_14_14 = Embedding(metadata['p5_14_14'], 10)(input7)
p5_14_14 = Reshape((10,))(p5_14_14)
print(K.shape(p5_14_14))

input8 = Input(shape=(1,),name='p5_4_input')
print(K.shape(input8))
p5_4 = Embedding(metadata['p5_4'], 10)(input8)
p5_4 = Reshape((10,))(p5_4)
print(K.shape(p5_4))

input9 = Input(shape=(1,),name='p6_4_input')
print(K.shape(input9))
p6_4 = Embedding(metadata['p6_4'], 10)(input9)
p6_4 = Reshape((10,))(p6_4)
print(K.shape(p6_4))

input10 = Input(shape=(18,), name='coords_input')
print(K.shape(input10))
coords = Dense(1,)(input10)
print(K.shape(coords))

input11 = Input(shape=(1,), name='distance_input')
print(K.shape(input11))
distance = Dense(1)(input11)
print(K.shape(distance))

input12 = Input(shape=(1,), name='duration_input')
print(K.shape(input12))
duration = Dense(1)(input12)
print(K.shape(duration))

data_merge = concatenate([p5_13, sexo, edad, estrato, p5_14_01, p5_14_02, p5_14_05, p5_14_14, p5_4, p6_4, coords, distance, duration])
layer_1 = Dense(500, activation="relu")(data_merge)
layer_2 = Dense(len(clusters),activation='softmax')(layer_1)
cast_clusters = K.cast_to_floatx(clusters)
print(cast_clusters.shape)
def destination(probabilities):
    return tf.matmul(probabilities, cast_clusters)
output_layer = Activation(destination)(layer_2)
model = Model(inputs=[input0, input1, input2, input3, input4, input5, input6, input7, input8, input9, input10, input11, input12], outputs=output_layer, name='data')

def tf_haversine(latlon1, latlon2):
    """
    Tensorflow version of the Haversine function to calculate distances between two sets of points.
    """
    lat1 = latlon1[:, 0]
    lon1 = latlon1[:, 1]
    lat2 = latlon2[:, 0]
    lon2 = latlon2[:, 1]

    REarth = 6371
    lat = tf.abs(lat1 - lat2) * np.pi / 180
    lon = tf.abs(lon1 - lon2) * np.pi / 180
    lat1 = lat1 * np.pi / 180
    lat2 = lat2 * np.pi / 180
    a = tf.sin(lat / 2) * tf.sin(lat / 2) + tf.cos(lat1) * tf.cos(lat2) * tf.sin(lon / 2) * tf.sin(lon / 2)
    d = 2 * tf_atan2(tf.sqrt(a), tf.sqrt(1 - a))
    return REarth * d


def tf_atan2(y, x):
    """
    Tensorflow doesn't have an Atan2 function (at least not yet, see: https://github.com/tensorflow/tensorflow/issues/6095).
    So we define it here ourselves.
    """
    angle = tf.where(tf.greater(x,0.0), tf.atan(y/x), tf.zeros_like(x))
    angle = tf.where(tf.logical_and(tf.less(x,0.0),  tf.greater_equal(y,0.0)), tf.atan(y/x) + np.pi, angle)
    angle = tf.where(tf.logical_and(tf.less(x,0.0),  tf.less(y,0.0)), tf.atan(y/x) - np.pi, angle)
    angle = tf.where(tf.logical_and(tf.equal(x,0.0), tf.greater(y,0.0)), 0.5*np.pi * tf.ones_like(x), angle)
    angle = tf.where(tf.logical_and(tf.equal(x,0.0), tf.less(y,0.0)), -0.5*np.pi * tf.ones_like(x), angle)
    angle = tf.where(tf.logical_and(tf.equal(x,0.0), tf.equal(y,0.0)), np.nan * tf.zeros_like(x), angle)
    return angle

optimizer = SGD(learning_rate=0.01, decay=1e-6, momentum=0.9, nesterov=True, clipvalue=1.)
model.compile(loss=tf_haversine, optimizer=optimizer)
```

```python
import math
import pandas as pd
import tensorflow as tf
import matplotlib.pyplot as plt
from tensorflow.keras import Model
from tensorflow.keras import Sequential
from sklearn.preprocessing import StandardScaler
from tensorflow.keras.optimizers import Adam
from tensorflow.keras.layers import Dense, Dropout
from sklearn.model_selection import train_test_split
from tensorflow.keras.losses import MeanSquaredLogarithmicError
```

```python
def scale_dataset(X_train,X_test):
    
    ''' 
    Standard Scale test and train data
    Z - Score normalization'''
    
    standard_scaler = StandardScaler()
    x_train_scaled = pd.DataFrame(standard_scaler.fit_transform(X_train), columns = X_train.columns)
    x_test_scaled = pd.DataFrame(standard_scaler.transform(X_test),columns = X_test.columns)
    
    return x_train_scaled, x_test_scaled

x_train_scaled, x_test_scaled = scale_dataset(X_train, X_test)
```

```python
hidden_units1 = 160
hidden_units2 = 480
hidden_units3 = 256
learning_rate = 0.01
# Creating model using the Sequential in tensorflow
def build_model_using_sequential():
    model = Sequential([
    Dense(hidden_units1, kernel_initializer='normal', activation='relu'),
    Dropout(0.2),
    Dense(hidden_units2, kernel_initializer='normal', activation='relu'),
    Dropout(0.2),
    Dense(hidden_units3, kernel_initializer='normal', activation='relu'),
    Dense(1, kernel_initializer='normal', activation='linear')
    ])
    return model
# build the model
model = build_model_using_sequential()
```

```python
# loss function
msle = MeanSquaredLogarithmicError()
model.compile(
    loss=msle, 
    optimizer=Adam(learning_rate=learning_rate), 
    metrics=[msle]
)
# train the model
history = model.fit(
    x_train_scaled.values, 
    y_train.values, 
    epochs=50, 
    batch_size=64,
    validation_split=0.2
)
```
